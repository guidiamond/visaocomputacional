Definido o ponto de aquisição, fazemos a leitura da imagem e eventuais processamentos. No caso da imagem abaixo, que está no sistema de cores BGR (RGB invertido), fizemos sua conversão para níveis de cinza (grayscale).

```python
img=cv2.imread('lena.jpg')
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

Finalmente, a imagem pode ser exibida... eventualmente, pode ser necessário o uso de **mapas de cores (colormaps)** específicos.

```python
# fig são as janelas GUI (começam em 1 em vez de 0)
# Axes são muito semelhantes aos subplots, mas permitem a colocação de plotagens em qualquer local na figura.
fig,ax=plt.subplots(ncols=2) # Organiza os plots em uma malha regular (ncols=1,nrows=1)
# ... serve apenas para objetos do tipo Ellipsis (objetos que implementam __getitem__ como o numpy e outras bibliotecas)
# o ... é util geralmente para fazer slicing em várias dimensões.

ax[0].imshow(img[...,::-1])  # BGR para RGB: em todas as dimensões, inverta a ordem dos arrays
ax[0].set_title('Imagem Original')
ax[1].imshow(gray_img,cmap=plt.cm.gray)
ax[1].set_title('Imagem em Níveis de Cinza')
plt.show()
```

Converter a imagem img para binária (branco e preto), armazená-la numa nova variável e fazer sua exibição.


```python
#faça sua implementação aqui
#Referência: https://docs.opencv.org/4.5.2/d6/d00/tutorial_py_root.html

# Threshold é uma técnica de definir valor para os pixels baseado em um valor de corte(threshold) fornecido
ret, blackAndWhiteImage = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY) 
# O cmap (Color Map) é um array de cores que são usados para mapear o dado de um pixel (representados como índices em uma tabela de cores)
# como o valor atual da cor. (lookup table)

# Explicação:

"""
Some graphic displays, especially older ones, cannot show an arbitrary color for every pixel.
Instead, each pixel is stored as a small number that's used as an index into the color table.
For example, it was common for VGA, in some modes, to be able to display up to 256 unique colors at one time.
But since the screen had many more than 256 pixels, you couldn't set an arbitrary RGB value for each of them.
Instead, you'd store a single byte per pixel, which would be used to look up the actual color in the color table.
"""

# O RGB ainda represnta cores com uma color table, em vez de codificar os pixels como cores em algum espectro de cores, por isso sempre que
# você for mostrar uma imagem é nessário usar o cmap dela
plt.imshow(blackAndWhiteImage, cmap=plt.cm.gray)
```
